```python
import zipfile
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Extract the ZIP file
zip_path = r"C:\Users\Manik Prabhu\Downloads\API_SP.POP.TOTL_DS2_en_csv_v2_127006.zip"
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall("population_data")

# Step 2: Load the main population dataset
data_path = "population_data/API_SP.POP.TOTL_DS2_en_csv_v2_127006.csv"
df = pd.read_csv(data_path, skiprows=4)

# Step 3: Clean and extract population data for the year 2023
df_2023 = df[['Country Name', '2023']].dropna()
populations_2023 = df_2023['2023']

# Step 4: Plot histogram with custom styling
plt.figure(figsize=(12, 7))
plt.hist(populations_2023, bins=40, color='#4B8BBE', edgecolor='white', alpha=0.9, linewidth=1.2)

# Customization
plt.title(' Population Distribution Across Countries (2023)', fontsize=16, fontweight='bold', color='#333333')
plt.xlabel('Population', fontsize=13)
plt.ylabel('Number of Countries', fontsize=13)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)


# Show the chart
plt.show()
```
